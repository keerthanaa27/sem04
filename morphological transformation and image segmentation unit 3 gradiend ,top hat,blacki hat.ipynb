# morphological transformation and image segmentation

```python
import cv2 as cv
import matplotlib.pyplot as plt 
import numpy as np
```

# dilation

```python
img = cv.imread('C:/Users/SN115/Pictures/yk/j.png',0)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,225,cv.THRESH_BINARY)
kernal = np.ones((3,3),np.uint8)
dilation = cv.dilate(mask,kernal,iterations=1)
erode = cv.erode(mask,kernal,iterations=1)
titles = ['images','mask','dilate','erode']
images = [img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

# morpholigical opening

```python
open = cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

# display morphologial opening

```python
cv.imshow("img",img)
cv.imshow("morph open",open)
cv.waitKey(0)
cv.destroyAllWindows()
```

# diplay morphological gradient

```python
gradient = cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

```python
cv.imshow("ORIGINAL",img)
cv.imshow("MORPH GRADIENT",gradient)
cv.waitKey(0)
cv.destroyAllWindows()
```

# display blackhat

```python
blackhat= cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

```python
cv.imshow("ORIGINAL",img)
cv.imshow("BLACK HAT",blackhat)
cv.waitKey(0)
cv.destroyAllWindows()
```
